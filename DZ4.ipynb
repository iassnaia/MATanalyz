Решение 2. Построить график функции:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Определим функции
def func_a(x):
    return x**2 + 4*x + 3

def func_b(x):
    return -2 * np.sin(3 * x)

def func_c(x):
    return x - np.floor(x) - 0.5
```

```python
# Создадим массивы значений x
x_values = np.linspace(-10, 10, 400)
x_values_mod = np.linspace(-10, 10, 1000)  # для функции с дробной частью
```

```python
# Построим графики
plt.figure(figsize=(18, 6))

# График для y = x^2 + 4x + 3
plt.subplot(1, 3, 1)
plt.plot(x_values, func_a(x_values), label=r'$y = x^2 + 4x + 3$')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$y = x^2 + 4x + 3$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
```

```python
# График для y = -2 \sin 3x
plt.subplot(1, 3, 2)
plt.plot(x_values, func_b(x_values), label=r'$y = -2 \sin 3x$', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$y = -2 \sin 3x$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
```

```python
# График для y = {x} - 1/2
plt.subplot(1, 3, 3)
plt.plot(x_values_mod, func_c(x_values_mod), label=r'$y = \{x\} - \frac{1}{2}$', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$y = \{x\} - \frac{1}{2}$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

Решение 3. Найти пределы:

```python
# Определим переменные
x = sp.symbols('x')
```

```python
# 1) lim (x -> 5) (x^2 - 6x + 5) / (x^2 - 25)
limit_1 = sp.limit((x**2 - 6*x + 5) / (x**2 - 25), x, 5)
print("1) lim_{x -> 5} (x^2 - 6x + 5) / (x^2 - 25) =", limit_1)
```

```python
# 2) lim (x -> 1) (x^3 + x^2 + 2) / (x^3 + 1)
limit_2 = sp.limit((x**3 + x**2 + 2) / (x**3 + 1), x, 1)
print("2) lim_{x -> 1} (x^3 + x^2 + 2) / (x^3 + 1) =", limit_2)
```

```python
# 3) lim (x -> 3) (sqrt(2x + 3) - 3) / (x - 1)
limit_3 = sp.limit((sp.sqrt(2*x + 3) - 3) / (x - 3), x, 3)
print("3) lim_{x -> 3} (sqrt(2x + 3) - 3) / (x - 3) =", limit_3)
```

```python
# 4) lim (x -> 0) (1 - cos(x)) / x^2
limit_4 = sp.limit((1 - sp.cos(x)) / x**2, x, 0)
print("4) lim_{x -> 0} (1 - cos(x)) / x^2 =", limit_4)
```

```python
# 5) lim (x -> 0) x * cot(x)
limit_5 = sp.limit(x * sp.cot(x), x, 0)
print("5) lim_{x -> 0} x * cot(x) =", limit_5)
```

```python
# 6) lim (x -> 0) 3-я корень из (1 + 3x)
limit_6 = sp.limit(sp.root(1 + 3*x, 5), x, 0)
print("6) lim_{x -> 0} 5-я корень из (1 + 3x) =", limit_6)
```

```python
# 7) lim (x -> 0) ((3 + 5x) / (3 + 2x))^2
limit_7 = sp.limit(((3 + 5*x) / (3 + 2*x))**2, x, 0)
print("7) lim_{x -> 0} ((3 + 5x) / (3 + 2x))^2 =", limit_7)
```
